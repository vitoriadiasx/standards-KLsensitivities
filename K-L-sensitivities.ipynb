```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from matplotlib.offsetbox import AnchoredText
```

```python
K_data = pd.read_csv("K-data.csv", header=0, decimal=",")
L_data = pd.read_csv("L-data.csv", header=0, decimal=",")
```

```python
def func(x, a, b, c, d, e, f):
    return a*x**5 + b*x**4 + c*x**3 + d*x**2 + e*x + f
```

```python
# K SENSITIVITY

K_popt, K_pcov = curve_fit(func, K_data.X, K_data.Y, p0=[1,1,1,1,1,1])

K_y_fit = func(K_data.X, *K_popt)

parameters_text = AnchoredText('a = {:.2E} \n b = {:.2E} \n c = {:.2E} \n d = {:.2E} \n e = {:.2E} \n f = {:.2E}'.format(*K_popt), loc = 'upper right')

fig, ax = plt.subplots()

ax.errorbar(K_data.X, K_data.Y, yerr = K_data.sY, fmt = '.')
ax.plot(K_data.X, K_y_fit, label = "polynomial fit")
ax.add_artist(parameters_text)

plt.title("Número Atômico X Sensibilidade K")
plt.ylabel("Sensibilidade K")
plt.xlabel("Número Atômico (Z)")

plt.savefig("K_sensitivity.png", dpi=300)

print("a =", K_popt[0], "#", K_pcov[0,0]**0.5)
print("b =", K_popt[1], "#", K_pcov[1,1]**0.5)
print("c =", K_popt[2], "#", K_pcov[2,2]**0.5)
print("d =", K_popt[3], "#", K_pcov[3,3]**0.5)
print("e =", K_popt[4], "#", K_pcov[4,4]**0.5)
print("f =", K_popt[5], "#", K_pcov[5,5]**0.5)
```

```python
# L SENSITIVITY

L_popt, L_pcov = curve_fit(func, L_data.X, L_data.Y, p0=[1,1,1,1,1,1])

L_y_fit = func(L_data.X, *L_popt)

parameters_text = AnchoredText('a = {:.2E} \n b = {:.2E} \n c = {:.2E} \n d = {:.2E} \n e = {:.2E} \n f = {:.2E}'.format(*L_popt), loc = 'upper right')

fig, ax = plt.subplots()

ax.errorbar(L_data.X, L_data.Y, yerr = L_data.sY, fmt = '.')
ax.plot(L_data.X, L_y_fit, label = "polynomial fit")
ax.add_artist(parameters_text)

plt.title("Número Atômico X Sensibilidade L")
plt.ylabel("Sensibilidade L")
plt.xlabel("Número Atômico (Z)")

plt.savefig("L_sensitivity.png", dpi=300)

print("a =", L_popt[0], "#", L_pcov[0,0]**0.5)
print("b =", L_popt[1], "#", L_pcov[1,1]**0.5)
print("c =", L_popt[2], "#", L_pcov[2,2]**0.5)
print("d =", L_popt[3], "#", L_pcov[3,3]**0.5)
print("e =", L_popt[4], "#", L_pcov[4,4]**0.5)
print("f =", L_popt[5], "#", L_pcov[5,5]**0.5)
```
